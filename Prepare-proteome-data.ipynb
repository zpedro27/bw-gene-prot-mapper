```python
import pandas as pd
```

```python
df_raw = pd.read_excel("data/[Schmidt2016]41587_2016_BFnbt3418_MOESM18_ESM.xlsx",
                       sheet_name="Table S6", usecols="A:AC", skiprows=2)
df_raw.head()
```

```python
df_filt = df_raw.drop(["Description", "Peptides.used.for.quantitation", "Confidence.score", "Molecular weight (Da)", "Dataset"], axis=1)
df_filt.head()
```

```python
df = pd.melt(df_filt, id_vars=["Gene", "Uniprot Accession"], value_name="ProteinAbundance", var_name="condition")
df = df.sort_values(by="Gene")
df
```

```python
df.to_csv("data/schmidt2016_s6.csv")
```
